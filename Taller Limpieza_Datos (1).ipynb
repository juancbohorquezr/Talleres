### Ejercicio Duración del contrato
Teniendo en cuenta que la duración de los contratos tiene métricas diferentes, encuentre una estrategia para crear una nueva variable en la cuál se encuentre parametrizada la variable de duración del contrato.

```python
df_SECOP['Unidad de Duracion'].value_counts()
```

```python
df_SECOP[['Unidad de Duracion','Duracion']].sample(20)
```

Se puede observar como la duración del contrato está dada en Dias, Meses, Años y ND por lo cual intentaremos crear una sola medida calculada en días.  

```python
df_TotalDias = df_SECOP.iloc[ : , 15:17]
df_TotalDias
```

```python
largo = df_TotalDias.size
```

```python
j = 0
for i in range(largo):
    if df_TotalDias.iloc[i,1] == 'Meses':
        df_TotalDias.iloc[j,0] = df_TotalDias.iloc[i,0]*30.4375
        df_TotalDias.iloc[j,1] = 'Días'
    elif df_TotalDias.iloc[i,1] == 'Años':
        df_TotalDias.iloc[j,0] = df_TotalDias.iloc[i,0]*365.25
        df_TotalDias.iloc[j,1] = 'Días'
    elif df_TotalDias.iloc[i,1] == 'ND':
        df_TotalDias.iloc[j,0]
        df_TotalDias.iloc[j,1] = 'Días'
    elif df_TotalDias.iloc[i,1] == 'Días':
        df_TotalDias.iloc[j,0] 
        df_TotalDias.iloc[j,1] = 'Días'
    j = j+1
```

```python
df_TotalDias[1:5]
```

Aquí podemos ver la columna que unifica una sola medida de la Duracion del contrato calculada en días. Aunque computacionalmente es desgastante correr este "for".
